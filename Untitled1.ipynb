```python
# Section 1: Setup, Controls, and Data Generation

import numpy as np
import random
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt

# Reproducibility Control
# We set random seeds for Python's `random`, NumPy, and TensorFlow.
# This ensures that every time the script is run, the same "random" data is generated
# and the model weights are initialized in the same way. This makes our results
# perfectly reproducible, which is crucial for experiments.
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
tf.random.set_seed(SEED)

#  Constants
# Define the building blocks of our mathematical language.
OPERATORS = ['+', '-', '*', '/']
IDENTIFIERS = list('abcde') # Variables or operands.
SPECIAL_TOKENS = ['PAD', 'SOS', 'EOS'] # Special tokens for sequence modeling.
# PAD: Padding token to make all sequences the same length.
# SOS: "Start of Sequence" token to signal the decoder to start generating.
# EOS: "End of Sequence" token to signal the end of a sequence.

# The full vocabulary of our language.
SYMBOLS = ['(', ')'] + OPERATORS
VOCAB = SPECIAL_TOKENS + SYMBOLS + IDENTIFIERS

# Create mappings from tokens to integer IDs and vice-versa.
# Neural networks work with numbers, not characters.
token_to_id = {tok: i for i, tok in enumerate(VOCAB)}
id_to_token = {i: tok for tok, i in token_to_id.items()}

# Store important vocabulary constants.
VOCAB_SIZE = len(VOCAB)
PAD_ID = token_to_id['PAD']
EOS_ID = token_to_id['EOS']
SOS_ID = token_to_id['SOS']

# Define the complexity of the expressions to generate.
MAX_DEPTH = 3
# A safe maximum length for sequences to ensure all expressions fit.
MAX_LEN = 30

# Expression Generation Functions

def generate_infix_expression(max_depth):
    """
    Recursively generates a random, fully-parenthesized infix expression.
    - max_depth: Controls the maximum nesting level of the expression.
    - The probability of terminating (generating a single identifier) is set to 0.2
      to encourage the creation of more complex expressions on average.
    """
    if max_depth == 0 or (max_depth > 0 and random.random() < 0.2):
        return random.choice(IDENTIFIERS)
    else:
        left = generate_infix_expression(max_depth - 1)
        right = generate_infix_expression(max_depth - 1)
        op = random.choice(OPERATORS)
        # Always wrap operations in parentheses to remove ambiguity.
        return f'({left}{op}{right})'

def tokenize(expr):
    """Converts a string expression into a list of character tokens."""
    return [c for c in expr if c in token_to_id]

def infix_to_postfix(tokens):
    """
    Converts a list of infix tokens to postfix using the Shunting-yard algorithm.
    This provides the "ground truth" (correct answer) for training the model.
    """
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2}
    output, stack = [], []
    for token in tokens:
        if token in IDENTIFIERS:
            output.append(token)
        elif token in OPERATORS:
            while stack and stack[-1] in OPERATORS and precedence.get(stack[-1], 0) >= precedence.get(token, 0):
                output.append(stack.pop())
            stack.append(token)
        elif token == '(':
            stack.append(token)
        elif token == ')':
            while stack and stack[-1] != '(':
                output.append(stack.pop())
            if stack and stack[-1] == '(':
                stack.pop() # Discard the left parenthesis.
    while stack:
        output.append(stack.pop())
    return output

def encode(tokens, max_len=MAX_LEN):
    """
    Converts a list of tokens into a fixed-length numerical sequence.
    - Maps each token to its ID.
    - Appends an EOS token.
    - Pads the sequence with PAD tokens to reach `max_len`.
    """
    ids = [token_to_id[t] for t in tokens] + [EOS_ID]
    padding_needed = max_len - len(ids)
    return ids + [PAD_ID] * padding_needed

def decode_sequence(token_ids, id_to_token):
    """Converts a numerical sequence back into a human-readable string."""
    tokens = []
    for token_id in token_ids:
        token = id_to_token.get(int(token_id), '?')
        if token == 'EOS':
            break
        if token != 'PAD':
            tokens.append(token)
    return "".join(tokens)

def generate_dataset(n, max_depth=MAX_DEPTH):
    """
    Generates a full dataset of `n` unique infix-postfix pairs.
    The dataset is stored in memory as NumPy arrays.
    """
    X, Y = [], [] # X stores infix (input), Y stores postfix (target)
    seen = set() # Used to ensure all generated expressions are unique.
    while len(X) < n:
        expr = generate_infix_expression(max_depth)
        if expr in seen:
            continue
        seen.add(expr)

        infix_tokens = tokenize(expr)
        postfix_tokens = infix_to_postfix(infix_tokens)
        X.append(encode(infix_tokens, MAX_LEN))
        Y.append(encode(postfix_tokens, MAX_LEN))

    return np.array(X), np.array(Y)

def shift_right(seqs):
    """
    Prepares the target sequences for the decoder during training (teacher forcing).
    - It shifts the entire sequence one step to the right.
    - It prepends the SOS token at the beginning.
    Example: [a, b, +] becomes [SOS, a, b]. This way, the decoder learns
    to predict the next token given the previous correct token.
    """
    shifted = np.zeros_like(seqs)
    shifted[:, 1:] = seqs[:, :-1]
    shifted[:, 0] = SOS_ID
    return shifted
```

```python
# Section 2: Dataset Creation

print("Generating datasets (with a fixed seed for reproducibility)...")
# Training dataset
X_train, Y_train = generate_dataset(20000, max_depth=MAX_DEPTH)
decoder_input_train = shift_right(Y_train)

# Validation dataset (used to monitor performance and prevent overfitting)
X_val, Y_val = generate_dataset(2000, max_depth=MAX_DEPTH)
decoder_input_val = shift_right(Y_val)
print("Datasets generated.")
```

```python
# Section 3: Model Definition and Training with Controls

# Model Hyperparameters
embedding_dim = 128   # The size of the vector space for representing tokens.
latent_dim = 256      # The number of units in the LSTM layer (the model's "memory").
dropout_rate = 0.3    # The fraction of units to drop for regularization.

# Encoder
# The encoder's job is to read the input sequence and compress its meaning
# into a fixed-size context vector (the LSTM's final state).
encoder_inputs = layers.Input(shape=(None,), name='encoder_input')
# The Embedding layer turns integer IDs into dense vectors. `mask_zero=True` tells
# the model to ignore PAD tokens.
enc_emb = layers.Embedding(VOCAB_SIZE, embedding_dim, mask_zero=True)(encoder_inputs)
# The LSTM layer processes the sequence. `return_state=True` provides us with the
# final hidden state (state_h) and cell state (state_c).
encoder_lstm = layers.LSTM(latent_dim, return_state=True, name='encoder_lstm')
_, state_h, state_c = encoder_lstm(enc_emb)
# We discard the encoder's output sequence and only keep its final state.
encoder_states = [state_h, state_c] # This is the "context vector".

# Decoder
# The decoder's job is to take the context vector and generate the output sequence,
# one token at a time.
decoder_inputs = layers.Input(shape=(None,), name='decoder_input')
# We reuse the same embedding layer for the decoder.
dec_emb_layer = layers.Embedding(VOCAB_SIZE, embedding_dim, mask_zero=True)
dec_emb = dec_emb_layer(decoder_inputs)
# The decoder LSTM is initialized with the encoder's final state.
# `return_sequences=True` makes it output a token for every step in the sequence.
decoder_lstm = layers.LSTM(latent_dim, return_sequences=True, return_state=True, name='decoder_lstm')
decoder_outputs, _, _ = decoder_lstm(dec_emb, initial_state=encoder_states)

# A Dropout layer is added for regularization
# It randomly sets a fraction of input units to 0 at each update during training time,
# which helps prevent overfitting.
decoder_dropout = layers.Dropout(dropout_rate)(decoder_outputs)

# A Dense layer with a softmax activation predicts the probability of each token
# in the vocabulary for each position in the output sequence.
decoder_dense = layers.Dense(VOCAB_SIZE, activation='softmax', name='decoder_output')
decoder_outputs = decoder_dense(decoder_dropout)

# Full Model for Training
# This model takes both the encoder and decoder inputs and produces the final output.
model = models.Model([encoder_inputs, decoder_inputs], decoder_outputs)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Define the EarlyStopping Callback
# This is a crucial control to prevent overfitting. It monitors a chosen metric
# (here, validation loss) and stops the training if it doesn't improve for a
# specified number of epochs (`patience`).
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=3,
    verbose=1, # Print a message when training is stopped.
    restore_best_weights=True # Restore model weights from the epoch with the best value.
)

# Training with Controls
print("\nStarting model training with Early Stopping...")
history = model.fit(
    [X_train, decoder_input_train], # Model inputs
    Y_train, # Model targets
    batch_size=64,
    epochs=20, # We can set a higher number; EarlyStopping will likely stop it sooner.
    validation_data=([X_val, decoder_input_val], Y_val),
    callbacks=[early_stopping] # Pass the callback here.
)
print("Training complete.")

# Visualize Training Results
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
# Section 4: Inference Model and Decoding Function
# (This section will now use the improved, early-stopped model)

# For inference (prediction), we need to run the model step-by-step (autoregressively).
# We build separate models for the encoder and decoder to manage the state manually.

# 1. Encoder Model: Takes the input sequence and returns the context state.
encoder_model = models.Model(encoder_inputs, encoder_states)

# 2. Decoder Model: Takes the current target token and the previous state,
# and returns the prediction for the next token and the new state.
decoder_state_input_h = layers.Input(shape=(latent_dim,))
decoder_state_input_c = layers.Input(shape=(latent_dim,))
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

dec_emb2 = dec_emb_layer(decoder_inputs)
decoder_outputs2, state_h2, state_c2 = decoder_lstm(dec_emb2, initial_state=decoder_states_inputs)
decoder_states2 = [state_h2, state_c2]
# Note: For inference, we connect the dense layer directly to the LSTM output,
# as dropout should only be active during training.
decoder_outputs2 = decoder_dense(decoder_outputs2)

decoder_model = models.Model(
    [decoder_inputs] + decoder_states_inputs,
    [decoder_outputs2] + decoder_states2
)

def autoregressive_decode(model, encoder_input_data):
    """
    Generates a postfix translation for a given infix input, token by token.
    """
    # 1. Encode the input to get the initial context state.
    states_value = encoder_model.predict(np.array([encoder_input_data]), verbose=0)

    # 2. Start the output sequence with the SOS token.
    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = SOS_ID

    decoded_tokens = []

    # 3. Loop to generate tokens until EOS is predicted or max length is reached.
    for _ in range(MAX_LEN):
        # Predict the next token and get the new state.
        output_tokens, h, c = decoder_model.predict([target_seq] + states_value, verbose=0)

        # Choose the token with the highest probability.
        sampled_token_id = np.argmax(output_tokens[0, -1, :])

        # Stop if the end-of-sequence token is predicted.
        if sampled_token_id == EOS_ID:
            break

        decoded_tokens.append(sampled_token_id)

        # Prepare the input for the next step.
        target_seq = np.zeros((1, 1))
        target_seq[0, 0] = sampled_token_id
        states_value = [h, c]

    return decoded_tokens
```

```python
# Section 5: Final Evaluation

def prefix_accuracy_single(y_true, y_pred, id_to_token):
    """
    Calculates the "prefix accuracy" metric. It measures what fraction of the
    generated sequence matches the ground truth from the beginning.
    A score of 1.0 means a perfect match.
    """
    true_str = decode_sequence(y_true, id_to_token)
    pred_str = decode_sequence(y_pred, id_to_token)
    t_tokens, p_tokens = list(true_str), list(pred_str)
    max_len = max(len(t_tokens), len(p_tokens))
    if max_len == 0: return 1.0
    # Count how many tokens match from the start.
    match_len = 0
    for t, p in zip(t_tokens, p_tokens):
        if t == p:
            match_len += 1
        else:
            break
    return match_len / max_len

def test(no=20, rounds=10):
    """
    Performs a robust evaluation by running multiple rounds of tests on new, unseen data.
    Returns the mean and standard deviation of the scores across all rounds.
    """
    rscores = []
    print("\nStarting final evaluation...")
    for i in range(rounds):
        X_test, Y_test = generate_dataset(no, max_depth=MAX_DEPTH)
        scores = [prefix_accuracy_single(Y_test[j], autoregressive_decode(model, X_test[j]), id_to_token) for j in range(no)]
        round_mean = np.mean(scores)
        rscores.append(round_mean)
        print(f"Round {i+1}/{rounds}, Average Score: {round_mean:.4f}")
    return np.mean(rscores), np.std(rscores)

# Run the final test
mean_score, std_dev = test(no=20, rounds=10)
print("\n Final Evaluation Results")
print(f"Mean Prefix Accuracy Score: {mean_score:.4f}")
print(f"Standard Deviation: {std_dev:.4f}")
```
