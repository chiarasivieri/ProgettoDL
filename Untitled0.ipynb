```python
# ==============================================================================
# Sezione 1: Setup, Costanti e Generazione Dati
# ==============================================================================

import numpy as np
import random
import string
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

# -------------------- Costanti --------------------
OPERATORS = ['+', '-', '*', '/']
IDENTIFIERS = list('abcde')
SPECIAL_TOKENS = ['PAD', 'SOS', 'EOS']
SYMBOLS = ['(', ')'] + OPERATORS
VOCAB = SPECIAL_TOKENS + SYMBOLS + IDENTIFIERS

token_to_id = {tok: i for i, tok in enumerate(VOCAB)}
id_to_token = {i: tok for tok, i in token_to_id.items()}
VOCAB_SIZE = len(VOCAB)
PAD_ID = token_to_id['PAD']
EOS_ID = token_to_id['EOS']
SOS_ID = token_to_id['SOS']

MAX_DEPTH = 3
# Lunghezza massima per un'espressione infissa con parentesi e spazi e token EOS/PAD
MAX_LEN_INFIX = (2**(MAX_DEPTH + 1) - 1) * 2
# Lunghezza massima per un'espressione postfissa
MAX_LEN_POSTFIX = 2**(MAX_DEPTH + 1) - 1
MAX_LEN = max(MAX_LEN_INFIX, MAX_LEN_POSTFIX)

# -------------------- Funzioni di Generazione Espressioni --------------------
def generate_infix_expression(max_depth):
    if max_depth == 0 or (max_depth > 0 and random.random() < 0.3):
        return random.choice(IDENTIFIERS)
    else:
        left = generate_infix_expression(max_depth - 1)
        right = generate_infix_expression(max_depth - 1)
        op = random.choice(OPERATORS)
        return f'({left}{op}{right})'

def tokenize(expr):
    # Rimuove gli spazi e tokenizza
    return [c for c in expr.replace(' ', '') if c in token_to_id]

def infix_to_postfix(tokens):
    precedence = {'+': 1, '-': 1, '*': 2, '/': 2}
    output, stack = [], []
    for token in tokens:
        if token in IDENTIFIERS:
            output.append(token)
        elif token in OPERATORS:
            while stack and stack[-1] in OPERATORS and precedence[stack[-1]] >= precedence[token]:
                output.append(stack.pop())
            stack.append(token)
        elif token == '(':
            stack.append(token)
        elif token == ')':
            top_token = stack.pop()
            while top_token != '(':
                output.append(top_token)
                top_token = stack.pop()
    while stack:
        output.append(stack.pop())
    return output

def encode(tokens, max_len=MAX_LEN):
    ids = [token_to_id[t] for t in tokens] + [EOS_ID]
    padding_needed = max_len - len(ids)
    return ids + [PAD_ID] * padding_needed

def decode_sequence(token_ids, id_to_token):
    tokens = []
    for token_id in token_ids:
        token = id_to_token.get(int(token_id), '?')
        if token == 'EOS':
            break
        if token != 'PAD':
            tokens.append(token)
    return "".join(tokens)

def generate_dataset(n, max_depth=MAX_DEPTH):
    X, Y = [], []
    seen = set()
    while len(X) < n:
        expr = generate_infix_expression(max_depth)
        if expr in seen:
            continue
        seen.add(expr)

        infix_tokens = tokenize(expr)
        postfix_tokens = infix_to_postfix(infix_tokens)

        X.append(encode(infix_tokens, MAX_LEN))
        Y.append(encode(postfix_tokens, MAX_LEN))

    return np.array(X), np.array(Y)

def shift_right(seqs):
    shifted = np.zeros_like(seqs)
    shifted[:, 1:] = seqs[:, :-1]
    shifted[:, 0] = SOS_ID
    return shifted

# ==============================================================================
# Sezione 2: Creazione dei Dataset
# ==============================================================================

print("Generazione dei dataset in corso...")
X_train, Y_train = generate_dataset(20000, max_depth=MAX_DEPTH)
decoder_input_train = shift_right(Y_train)

X_val, Y_val = generate_dataset(2000, max_depth=MAX_DEPTH)
decoder_input_val = shift_right(Y_val)
print("Dataset generati.")

# Ispezione di un campione
i = np.random.randint(len(X_train))
print("\n--- Esempio dal dataset ---")
print(f"Infix (raw):      {decode_sequence(X_train[i], id_to_token)}")
print(f"Postfix (target): {decode_sequence(Y_train[i], id_to_token)}")
print(f"Decoder Input:    {decode_sequence(decoder_input_train[i], id_to_token)}")
print("---------------------------\n")


# ==============================================================================
# Sezione 3: Definizione e Addestramento del Modello
# ==============================================================================

# Definiamo un'architettura Encoder-Decoder con LSTM
embedding_dim = 128
latent_dim = 256

# --- Encoder ---
encoder_inputs = layers.Input(shape=(None,), name='encoder_input')
enc_emb = layers.Embedding(VOCAB_SIZE, embedding_dim, mask_zero=True)(encoder_inputs)
encoder_lstm = layers.LSTM(latent_dim, return_state=True, name='encoder_lstm')
_, state_h, state_c = encoder_lstm(enc_emb)
encoder_states = [state_h, state_c] # Il "contesto"

# --- Decoder ---
decoder_inputs = layers.Input(shape=(None,), name='decoder_input')
dec_emb_layer = layers.Embedding(VOCAB_SIZE, embedding_dim, mask_zero=True)
dec_emb = dec_emb_layer(decoder_inputs)
decoder_lstm = layers.LSTM(latent_dim, return_sequences=True, return_state=True, name='decoder_lstm')
decoder_outputs, _, _ = decoder_lstm(dec_emb, initial_state=encoder_states)
decoder_dense = layers.Dense(VOCAB_SIZE, activation='softmax', name='decoder_output')
decoder_outputs = decoder_dense(decoder_outputs)

# --- Modello Completo (per l'addestramento) ---
model = models.Model([encoder_inputs, decoder_inputs], decoder_outputs)

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# --- Addestramento ---
print("\nInizio addestramento del modello...")
history = model.fit(
    [X_train, decoder_input_train],
    Y_train,
    batch_size=64,
    epochs=15, # 10-15 epoche sono solitamente sufficienti per convergere
    validation_data=([X_val, decoder_input_val], Y_val)
)
print("Addestramento completato.")

# --- Visualizzazione Risultati Addestramento ---
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss del Modello')
plt.xlabel('Epoche')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy del Modello')
plt.xlabel('Epoche')
plt.ylabel('Accuracy')
plt.legend()
plt.show()


# ==============================================================================
# Sezione 4: Creazione del Modello di Inferenza e Funzione di Decodifica
# ==============================================================================

# Per l'inferenza, dobbiamo scomporre il modello per poterlo usare in modo autoregressivo

# 1. Modello Encoder
encoder_model = models.Model(encoder_inputs, encoder_states)

# 2. Modello Decoder
decoder_state_input_h = layers.Input(shape=(latent_dim,))
decoder_state_input_c = layers.Input(shape=(latent_dim,))
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

dec_emb2 = dec_emb_layer(decoder_inputs)

decoder_outputs2, state_h2, state_c2 = decoder_lstm(dec_emb2, initial_state=decoder_states_inputs)
decoder_states2 = [state_h2, state_c2]
decoder_outputs2 = decoder_dense(decoder_outputs2)

decoder_model = models.Model(
    [decoder_inputs] + decoder_states_inputs,
    [decoder_outputs2] + decoder_states2
)

# 3. Funzione di decodifica autoregressiva
def autoregressive_decode(model, encoder_input_data):
    # Usa l'encoder per ottenere il vettore di stato
    states_value = encoder_model.predict(np.array([encoder_input_data]), verbose=0)

    # Inizia la sequenza di output con il token SOS
    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = SOS_ID

    decoded_tokens = []

    for _ in range(MAX_LEN):
        output_tokens, h, c = decoder_model.predict([target_seq] + states_value, verbose=0)

        # Scegli il token con la probabilità più alta
        sampled_token_id = np.argmax(output_tokens[0, -1, :])

        if sampled_token_id == EOS_ID:
            break

        decoded_tokens.append(sampled_token_id)

        # Prepara l'input per il prossimo passo
        target_seq = np.zeros((1, 1))
        target_seq[0, 0] = sampled_token_id
        states_value = [h, c]

    return decoded_tokens

# ==============================================================================
# Sezione 5: Valutazione Finale
# ==============================================================================

def prefix_accuracy_single(y_true, y_pred, id_to_token):
    # Decodifica le sequenze ignorando i token speciali
    true_str = decode_sequence(y_true, id_to_token)
    pred_str = decode_sequence(y_pred, id_to_token)

    t_tokens = list(true_str)
    p_tokens = list(pred_str)

    # Calcola la lunghezza massima per la normalizzazione
    max_len = max(len(t_tokens), len(p_tokens))
    if max_len == 0:
        return 1.0 if len(t_tokens) == 0 and len(p_tokens) == 0 else 0.0

    # Calcola la lunghezza del prefisso comune
    match_len = 0
    for t, p in zip(t_tokens, p_tokens):
        if t == p:
            match_len += 1
        else:
            break

    score = match_len / max_len
    return score

def test(no=20, rounds=10):
    rscores = []
    print("\nInizio valutazione finale...")
    for i in range(rounds):
        X_test, Y_test = generate_dataset(no, max_depth=MAX_DEPTH)
        scores = []
        for j in range(no):
            encoder_input = X_test[j]
            # La funzione di decodifica non include SOS, quindi non serve rimuoverlo
            generated_ids = autoregressive_decode(model, encoder_input)
            score = prefix_accuracy_single(Y_test[j], generated_ids, id_to_token)
            scores.append(score)

        round_mean = np.mean(scores)
        rscores.append(round_mean)
        print(f"Round {i+1}/{rounds}, Score medio: {round_mean:.4f}")

    return np.mean(rscores), np.std(rscores)

# Eseguiamo il test
mean_score, std_dev = test(no=20, rounds=10)
print("\n--- Risultati Finali della Valutazione ---")
print(f"Punteggio medio di Prefix Accuracy: {mean_score:.4f}")
print(f"Deviazione Standard: {std_dev:.4f}")
print("------------------------------------------")

# ==============================================================================
# Sezione 6: Test su un singolo esempio
# ==============================================================================
print("\n--- Test su un singolo esempio nuovo ---")
X_test_single, Y_test_single = generate_dataset(1)
input_expr_ids = X_test_single[0]
target_expr_ids = Y_test_single[0]

predicted_ids = autoregressive_decode(model, input_expr_ids)

input_str = decode_sequence(input_expr_ids, id_to_token)
target_str = decode_sequence(target_expr_ids, id_to_token)
predicted_str = decode_sequence(predicted_ids, id_to_token)

print(f"Input Infix:    {input_str}")
print(f"Target Postfix: {target_str}")
print(f"Model Output:   {predicted_str}")

if target_str == predicted_str:
    print("Risultato: CORRETTO!")
else:
    print("Risultato: ERRATO.")
```
